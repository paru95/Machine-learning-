```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy (1).csv")
df.head(3)
```

```python
# df.isnull().sum()
```

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
# df.isnull().sum()
```

```python
df.shape
```

```python
x = df.drop(columns= 'has_covid')   #independent data
x.shape
```

```python
y = df['has_covid']    #target data
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train , y_test = train_test_split(x,
                                                   y,
                                                   test_size=0.2,
                                                   random_state=42)


```
